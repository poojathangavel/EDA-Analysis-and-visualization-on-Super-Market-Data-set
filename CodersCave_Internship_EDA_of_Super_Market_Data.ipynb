Plotly, Matplotlib.pyplot, Seaborn - These libraries are used for visualization of various plots.



```python
import pandas as pd
import plotly
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
d = pd.read_csv("/content/drive/MyDrive/data/Hackathon_Working_Data.csv")
```

This bar chart represents the sum of price of the products respective to the months. This plot shows that “month 2” has the highest sum of price compared to month1 and month3.

```python
plt.bar(d["MONTH"],d["PRICE"])
plt.xlabel ("MONTH")
plt.ylabel ("PRICE")
plt.title("PRICE PER MONTH")
plt.show()
```

Histogram represents the distribution of day count. Day 5 has the highest count of 1231, and day 31 has the least count of 429

```python
histogram = px.histogram(d, x ="DAY",title = "HISTOGRAM" )
histogram.show()
```

The heat map represents the correlation of day,bill amount, quantity, value and price

```python
corrolation = d.corr()
print(corrolation)
```

```python
sns.heatmap(corrolation,annot = True)
```

Scatter plot represents the distribution of bill amount respective to the day. Day 4 has the highest bill amount of 7292.

```python
scatter = px.scatter(d, x = "DAY", y = "BILL_AMT", title = "SCATTER PLOT")
scatter.show()
```

This pie chart represents the quantity sold respective to the store code. Store N1 has the highest percent of 36.6% and N6 has the least percent of 1.75%

```python
pie = px.pie(d,values="QTY",names = "STORECODE", title = "PIECHART")
pie.show()
```
